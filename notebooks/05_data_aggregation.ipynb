# Importing Libraries

```python
import pandas as pd
```

# Get Data

```python
df = pd.read_csv('../data/titanic.csv')
df.head()
```

# Basic Aggregation

```python
display(df['Survived'].mean().round(2))

display(df.groupby('Sex')['Survived'].count().round(2)) 

display('Surviving Male:' , df[(df['Sex'] == 'male') & (df['Survived'] == 1)].shape[0])
```

# Multi Aggregation

```python
df['Age'].agg(['min', 'max', 'mean', 'median']).round(2)
```

# Applying Statistics

```python
iqr_age = df['Age'].quantile(0.75) - df['Age'].quantile(0.25)
print('Interquartile Range of Age:', iqr_age.round(2))
```

```python
median_age = df['Age'].median()
mean_age = df['Age'].mean()

print('Median | Mean of Age')
print(median_age.round(2), '|', mean_age.round(2))
```

```python
std_age = df['Age'].std()
print('Standard Deviation of Age:', std_age.round(2))
```

# Practice

```python
# Survival Rate

survival_rate = df['Survived'].sum() / len(df) * 100
print('Survival Rate:', survival_rate.round(2) ,'%')
print('\n')

avg_survivor_age = df.loc[df['Survived'] == 1, 'Age'].mean()
avg_non_survivor_age = df.loc[df['Survived'] == 0, 'Age'].mean()

print('Average Age of Survivor', avg_survivor_age.round(2), 'years old')
print('Average Age of Non-Survivor', avg_non_survivor_age.round(2), 'years old')

print('\n')

avg_survivor_fare = df.loc[df['Survived'] == 1, 'Fare'].mean()
avg_non_survivor_fare = df.loc[df['Survived'] == 0, 'Fare'].mean()

print('Average Fare paid by Survivor', avg_survivor_fare.round(2), '$')
print('Average Fare paid by Non-Survivor', avg_non_survivor_fare.round(2), '$')
```
