# Import Libraries

```python
import pandas as pd
import numpy as np
```

# Create Data

```python
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4]
    , 'name': ['Alice', 'Bob', 'Charlie', 'David']
    , 'city': ['NYC','LA','Chicago','Houston']
})

df2 = pd.DataFrame({
    'order_id': [101, 102, 103, 104, 105]
    , 'customer_id' : [1,2,3,4,5]
    , 'amount': [100,150,200,75,300]
})

```

# Merge

Left Join

```python
left_merge = pd.merge(df, df2, on='customer_id', how='left')
display(left_merge)
```

Right Join

```python
right_merge = pd.merge(df, df2, on='customer_id', how='right')
display(right_merge)
```

Inner Join

```python
inner_merge = pd.merge(df, df2, on='customer_id', how='inner')
display(inner_merge)
```

Outer Join

```python
outer_merge = pd.merge(df, df2, on='customer_id', how='outer')
display(outer_merge)
```

# Concat

```python
df_concat = pd.concat([df, df2], axis=0, ignore_index=True, sort=False)
display(df_concat)
```

```python
df_concat = pd.concat([df, df2], axis=1, ignore_index=True, sort=False)
display(df_concat)
```

# Join

```python
df_join = df.join(df2.set_index('customer_id'), on='customer_id', how='right')
display(df_join)
```
