# Import Libraries

```python
import numpy as np
import pandas as pd
```

# Pivot

```python
df_sales_long = pd.DataFrame({
    'date':['2024-01-01', '2024-01-01', '2024-01-01','2024-01-02', '2024-01-02', '2024-01-02','2024-01-03', '2024-01-03', '2024-01-03']
    ,'product' : ['Laptop', 'Mouse', 'Keyboard','Laptop', 'Mouse', 'Keyboard','Laptop', 'Mouse', 'Keyboard']
    ,'sales' : [1200, 25, 75, 1500, 30, 80, 1100, 20, 70]
    ,'quantity' : [2, 5, 3, 3, 6, 4, 2, 4, 3]
})

display(df_sales_long)
```

```python
df_sales_long.pivot(index='date', columns='product', values='sales')
```

```python
df_sales_long.pivot_table(values='sales',index='date',columns='product',aggfunc='mean')
```

# Melt

```python
sales_wide = pd.DataFrame({
      'date': ['2024-01-01', '2024-01-02', '2024-01-03'],
      'Laptop': [1200, 1500, 1100],
      'Mouse': [25, 30, 20],
      'Keyboard': [75, 80, 70]
  })
sales_wide.head()
```

```python
pd.melt(sales_wide, id_vars=['date'], value_vars=['Laptop', 'Mouse', 'Keyboard'], var_name='product', value_name='sales')
```

# Advanced Pivot with Aggregation

```python
sales_detailed = pd.DataFrame({
      'date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02',
               '2024-01-01', '2024-01-02'],
      'region': ['North', 'South', 'North', 'South', 'North', 'South'],
      'product': ['Laptop', 'Laptop', 'Mouse', 'Mouse', 'Keyboard', 'Keyboard'],
      'sales': [1200, 1300, 25, 30, 75, 80],
      'quantity': [2, 2, 5, 6, 3, 4]
  })

sales_detailed.head()
```

```python
sales_detailed.pivot_table(index=['date'], aggfunc='sum', values=['sales'], columns='product')
```

```python
sales_detailed.pivot_table(index=['region'], aggfunc='mean', values=['sales'], columns='product')
```
