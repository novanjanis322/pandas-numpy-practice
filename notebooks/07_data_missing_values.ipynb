# Importing Libraries


```python
import pandas as pd
import numpy as np
```

# Create Sample Data

```python
data = {
    'A' : [1, 2, np.nan, 4, 5],
    'B' : [np.nan, 2, 3, np.nan, 5],
    'C' : [1, np.nan, np.nan, 4, 5],
    'D' : [1,2,3,4,5]
} 
df = pd.DataFrame(data)

df.head()
```

# Check Null

```python
display(
    df.isnull() \
    # .sum()
    )
df.info()
```

# Handle Missing Values

```python
df_cleaned = df.dropna(
    thresh=4
    ,axis = 1
)
df_cleaned.head()
```

```python
df_filled = df.fillna(
    0 \
    # df.mean() 
    # df.median()
    # method='ffill'
    # method='bfill'
    )
display(df_filled.head())

df_ffill = df.ffill()
display(df_ffill.head())


df_bfill = df.bfill()
df_bfill.head()
```

```python
df_interpolate = df.interpolate(
    method='linear'
)
display(df.head())
display(df_interpolate.head())
```
